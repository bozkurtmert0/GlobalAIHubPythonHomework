4. Yanlış tahmin edilen harfleri tutan liste  



**Kurallar** <p>
Yanlış tahmin edilen her harf için kaybetmeye daha da yaklaşırız. Toplam 11 yanlis cevapta oyun kaybedilir.

```
#kelimeyi rastgele almak için random
import random

isim = str(input("İsminizi giriniz :"))
print("Hoşgeldiniz sevgili {}".format(isim)

#kelimeleri listeye ekleyebilirsiniz
kelimeler = ['telefon','bilgisayar']
kelime_sirasi = random.randint(0,len(kelimeler)-1) #random kelime seçmek için
kelime = kelimeler[kelime_sirasi] #secılen kelımeyı değişkene atadık
kelime_harfleri = list(kelime) #kelımeyı liste haline getirdik 
harfsayisi = len(kelime)
#yanlış tahmin haklarımızı tutması için değişken 
skor = 0
bilinenler = [] #bulunan harfleri tutması için liste

yanlis_harfler = [] #tahmin edilen yanlış harfleri tutan

#oyun için basit bir arayüz 
print("------------------ Hangman Oyununa Hosgeldiniz  -----------------------------------") 
print("***Kurallar***")
print('10 yanlis harf tahmini hakkiniz var ,11. yanlis tahminde adam  daragacacini boylar')
print('Harf tahminlerinizi ard arda giriniz')
print("-----------------------------------------------------------------------------------")


#oyun için basit algoritma
while skor < 11 :
  
  print("***********************************")
  print("Kelimemiz {} harfli bir kelimedir".format(harfsayisi))
  harf = str(input("Tahmin harfini giriniz :"))
  print('Kalan tahmin hakkiniz {}'.format(11-skor))
  
  if harf in kelime_harfleri: #eger harf kelimemizde bulunuyorsa
    
    if harf in bilinenler: #eger harf aynı zamanda bilinen harflerin içindeyse
      print('Bu harfi daha once buldunuz')
      print("****************************************")
      skor -=1

    else: #eger harf bilinen harflerde yoksa ve kelimemizde bulunuyorsa
      print('Dogru harf tahmini')
      bilinenler.append(harf)
      print("****************************************")
      skor-=1

  else :
    print("Yanlış Tahmin")
    print("****************************************")
    yanlis_harfler.append(harf)

  skor+=1
  print("Buldugun harfler{}".format(bilinenler))
  print("Yanlış Tahmin edilen harfler {}".format(yanlis_harfler))

print('Kaybettiniz ,tahmin edemediginiz kelime {}'.format(kelime))
```
